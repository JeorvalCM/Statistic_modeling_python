```python
import numpy as np
from matplotlib import pyplot as plt
import math
import seaborn as sns
```

```python
np.random.seed(1)
N = 20
X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)),
                   np.random.normal(5, 1, int(0.7*N))))[:, np.newaxis]
X_plot = np.linspace(-5, 10, 1000)[:, np.newaxis]
bins = np.linspace(-5, 10, 10)
```

# Tarea

```python
def biweight(x0, h, x):
    delta = abs(x0-x)
    if delta < h:
        return (1-(delta/h)**2)**2
    else:
        return 0
```

```python
bandwidth = 1

Y = []
for x_val in X_plot:
    y_val = 0
    for x_dist in X:
        y_val += biweight(x_dist, bandwidth, x_val)
    y_val /= (bandwidth * len(X_plot))
    Y.append(y_val) 
```

```python
plt.figure(figsize = [8,4], dpi = 100)
plt.fill(X_plot[:,0], Y)
plt.xlim(-4,9)
plt.show()
```

```python
def triweight(x0, h, x):
    delta = abs(x0-x)
    if delta < h:
        return (1-(delta/h)**2)**3
    else:
        return 0
```

```python
bandwidth = 1

Y = []
for x_val in X_plot:
    y_val = 0
    for x_dist in X:
        y_val += triweight(x_dist, bandwidth, x_val)
    y_val /= (bandwidth * len(X_plot))
    Y.append(y_val) 
```

```python
plt.figure(figsize = [8,4], dpi = 100)
plt.fill(X_plot[:,0], Y)
plt.xlim(-4,9)
plt.show()
```

```python
def tricube(x0, h, x):
    delta = abs(x0-x)
    if delta < h:
        return (1-abs((delta/h)**3))**3
    else:
        return 0
```

```python
bandwidth = 1

Y = []
for x_val in X_plot:
    y_val = 0
    for x_dist in X:
        y_val += tricube(x_dist, bandwidth, x_val)
    y_val /= (bandwidth * len(X_plot))
    Y.append(y_val) 
```

```python
plt.figure(figsize = [8,4], dpi = 100)
plt.fill(X_plot[:,0], Y)
plt.xlim(-4,9)
plt.show()
```

```python
def logistic(x0, h, x):
    delta = abs(x0-x)
    return 1/(math.exp(delta/h)+2+math.exp(-delta/h))
```

```python
bandwidth = 1

Y = []
for x_val in X_plot:
    y_val = 0
    for x_dist in X:
        y_val += logistic(x_dist, bandwidth, x_val)
    y_val /= (bandwidth * len(X_plot))
    Y.append(y_val) 
```

```python
plt.figure(figsize = [8,4], dpi = 100)
plt.fill(X_plot[:,0], Y)
plt.xlim(-4,9)
plt.show()
```
